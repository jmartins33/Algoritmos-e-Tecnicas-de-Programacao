```python
import random

def menu():
    print(""" Gestão de lista de inteiros
(1) Introduza a lista à mão
(2) Gerar lista com números aleatórios
(3) Identificar o maior número da lista
(4) Bubble Sort
...
(0) Sair """)

def criarListaM():
    n=int(input("Introduza o número de elementos da lista:"))
    lista=[]
    for i in range (n):
        elem=int(input("Introduza o elemento"+  str(i+1) + ":"))
        lista.append(elem)
    return lista

def criarListaA():
    n=int(input("Introduza o número de elementos da lista:"))
    lista=[]
    for i in range (n):
        elem=random.randint(0,100)
        lista.append(elem)
    return lista

def maiorElemento(lista):
    maximo=lista[0]
    for i in range(len(lista)):
        if lista[i]>maximo:
            maximo=lista[i]
    return maximo
def bubbleSort(lista):
    trocas=-1
    while trocas !=0:
        trocas=0
        for i in range (len(lista)-1):
            if lista[i]>lista[i+1]:
                intermediario=lista[i]
                lista[i]=lista[i+1]
                lista[i+1]=intermediario
                trocas=trocas+1
    return lista
op=-3
lis=[]
while op!= 0:
    menu()
    op=int(input("Qual é a sua opção:"))
    if op == 1:
        lis=criarListaM()
        print(lis)
    elif op == 2:
        lis=criarListaA()
        print(lis)
    elif op == 3:
        if len(lis) !=0:
         maior=maiorElemento(lis)
         print(maior)
        else:
            op2=int(input("Você ainda não tem lista definida,faça uma com a opção 1 ou 2: "))
            if op2 == 1:
              lis=criarListaM()
              print(lis)
            elif op2 == 2:
              lis=criarListaA()
              print(lis)
    elif op == 4:
        if len(lis) !=0:
         ordenada=bubbleSort(lis)
         print(ordenada)
        else:
         op2=int(input("Você ainda não tem lista definida,faça uma com a opção 1 ou 2: "))
         if op2 == 1:
           lis=criarListaM()
           print(lis)
         elif op2 == 2:
           lis=criarListaA()
           print(lis)
    elif op == 0:
        print("Até à próxima..")
    else:
        print("Opção Inválida!")    
    
```

  A opção 1 e 2 funcionam de maneira semelhante, pois ambas começam com um lista vazia, pedem o número de elementos que nós queremos que a lista tenha, e depois enche a lista com esse número de elementos através do _lista.append[i]_ , a diferença está no critério de escolha desses elementos, na 1 o programa pede ao utilizador cada elemento para por na lista, no 2 através do módulo _random_ e da função _random.randint(1,100)_ conseguimos criar uma lista com _n_ elementos aleatórios.
  
  A opção 3 retorna o valor máximo da opção, usando uma variável __(maximo)__ que guardará um valor que definimos inicialmente com o primeiro da lista, esta variavél ira mudando ao longo que percorremos a lista, tomando o maior valor em cada comparação que faz dentro da lista, no fim ficando com o maior número da lista.
  
  A opção 4 permite-nos ordenar a lista, usando um método chamado de _Bubble Sort_ em que percorre a lista e faz trocas diretas entre um elemento e o elemento do índice seguinte, caso o primeiro seja maior, para que esta troca ocorra é necessário um __intermediário__ que toma o valor de um dos elementos antes da troca em si, para conseguirmos chegar à lista ordenada poderá ser necessário fazer este processo de percorrer a lista várias vezes, então com um contador __trocas__ ,que dá _reset_ no final de cada vez que percorre a lista, conseguimos perceber que quando este valor no final de correr a lista ainda estiver 0, a lista está ordenada.
