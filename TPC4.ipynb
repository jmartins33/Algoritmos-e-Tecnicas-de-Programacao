```python
import random

def criarFracao(numerador,denominador):
    if denominador!=0:
      return(numerador,denominador)
    else:
        print("Input inválido")
        criarFracao(numerador,denominador)

def verFracao(f):
    print(str(f[0])+ "/" + str(f[1]))

def mdc(n,d):
    if n<d:
        return mdc(d,n)
    elif n%d==0:
        return d
    else:
        return mdc(d,n%d)
    
def simplificarFracao(f):
    num,denom = f
    a = mdc(num,denom)
    return(num/a,denom/a)

def somarFrac(f1,f2):
    n1,d1=f1
    n2,d2=f2
    if d1==d2:
        return simplificarFracao(criarFracao(n1+n2,d1))
    else:
        return simplificarFracao(criarFracao(n1*d2+n2*d1,d1*d2))

def subtrairFrac(f1,f2):
    n1,d1=f1
    n2,d2=f2
    if d1==d2:
        return simplificarFracao(criarFracao(n1-n2,d1))
    else:
        return simplificarFracao(criarFracao(n1*d2-n2*d1,d1*d2))

def multiplicarFrac(f1,f2):
    n1,d1=f1
    n2,d2=f2
    return simplificarFracao(criarFracao(n1*n2,d1*d2))

def dividirFrac(f1,f2):
    n1,d1=f1
    n2,d2=f2
    return simplificarFracao(criarFracao(n1*d2,d1*n2))

def listaFracAl():
    n=int(input("Quantas frações quer que a lista tenha: "))
    l=[]
    for i in range (n):
        (a,b)=(random.randint(0,100),random.randint(1,100))
        l.append((a,b))
    return l

def somarLista(l):
   res=(0,1)
   for elem in l:
       res=somarFrac(res,elem)
   return res    
       
def maiorFrac(l):
    maior=(0,1)
    for elem in l:
        if elem[0]/elem[1] > maior[0]/maior[1]:
            maior=elem
    return maior

i=-2
l=[]
while i != 0:
    print(""" Opções disponíveis:
(1) Criar fração
(2) Simplificar fração
(3) Somar frações
(4) Subtrair frações
(5) Multiplicar frações
(6) Dividir frações
(7) Gerar lista de tamanho n, aleatória de frações
(8) Somar todos os elementos de uma lista de frações
(9) Maior fração de uma lista
(0) Sair """)
    i=int(input("Qual é a sua opção:"))
    if i == 1:
        n=int(input("Introduza o numerador da sua fração: "))
        d=int(input("Introduza o denominador da sua fração: "))
        fra=criarFracao(n,d)
        verFracao(fra)
    elif i == 2: 
        n=int(input("Introduza o numerador da sua fração: "))
        d=int(input("Introduza o denominador da sua fração: "))
        fra=criarFracao(n,d)
        fra=verFracao(simplificarFracao(fra))
    elif i == 3: 
        n=int(input("Introduza o numerador da sua fração: "))
        d=int(input("Introduza o denominador da sua fração: "))
        fra=criarFracao(n,d)
        n2=int(input("Introduza o numerador da sua segunda fração: "))
        d2=int(input("Introduza o denominador da sua segunda fração: "))
        fra2=criarFracao(n2,d2)
        fra=verFracao(somarFrac(fra,fra2))
    elif i==4:
        n=int(input("Introduza o numerador da sua fração: "))
        d=int(input("Introduza o denominador da sua fração: "))
        fra=criarFracao(n,d)
        n2=int(input("Introduza o numerador da sua segunda fração: "))
        d2=int(input("Introduza o denominador da sua segunda fração: "))
        fra2=criarFracao(n2,d2)
        fra=verFracao(subtrairFrac(fra,fra2))
    elif i==5:
        n=int(input("Introduza o numerador da sua fração: "))
        d=int(input("Introduza o denominador da sua fração: "))
        fra=criarFracao(n,d)
        n2=int(input("Introduza o numerador da sua segunda fração: "))
        d2=int(input("Introduza o denominador da sua segunda fração: "))
        fra2=criarFracao(n2,d2)
        fra=verFracao(multiplicarFrac(fra,fra2))
    elif i==6:
        n=int(input("Introduza o numerador da sua fração: "))
        d=int(input("Introduza o denominador da sua fração: "))
        fra=criarFracao(n,d)
        n2=int(input("Introduza o numerador da sua segunda fração: "))
        d2=int(input("Introduza o denominador da sua segunda fração: "))
        fra2=criarFracao(n2,d2)
        fra=verFracao(dividirFrac(fra,fra2))
    elif i==7:
        l=listaFracAl()
        print (l)
    elif i==8:
        if len(l) !=0:
         fra=somarLista(l)
        else:
         i2=int(input("Você ainda não tem lista definida,crie uma com a opção 7, quer prosseguir com a criação da lista? Escreva s se for o caso: "))
         if i2== "s":
             l=listaFracAl()
             print (l)
         
    elif i==9:
        if len(l) !=0:
         fra=verFracao(maiorFrac(l))
        else:
         i2=(input("Você ainda não tem lista definida,crie uma com a opção 7, quer prosseguir com a criação da lista? Escreva s se for o caso: "))
         if i2== "s":
             l=listaFracAl()
             print(l)
             fra=verFracao(maiorFrac(l))
        
    elif i==0:
        print("Até à próxima!")
    else:
        print("Opção inválida!")
```
